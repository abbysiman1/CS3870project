```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
flight_data  = pd.read_csv('satisfaction_2015.csv')
```

```python
flight_data.head()
```

```python
flight_data.shape
```

```python
missing_values_per_column = flight_data.isna().sum()
missing_values_per_column
```

```python
# Replacing the NA values of a column with the average of that column
average_column_arrival_delay = flight_data['Arrival Delay in Minutes'].mean()
average_column_arrival_delay
# Replace NaN values in column 'A' with the average of column 'A'
flight_data['Arrival Delay in Minutes'] = flight_data['Arrival Delay in Minutes'].fillna(average_column_arrival_delay)
flight_data.isna().sum()
```

```python
for column in flight_data.columns:
    unique_values = flight_data[column].unique()
    print(f"Unique values in column '{column}': {unique_values}")
```
